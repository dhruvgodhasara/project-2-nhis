### Outliers

```python
numerical_cols=dataset5.select_dtypes(include=[np.number])

# Calculate Q1, Q3, and IQR
Q1=numerical_cols.quantile(0.25)
Q3=numerical_cols.quantile(0.75)
IQR=Q3-Q1

# Define bounds
lower_bound= Q1 - 1.5 * IQR
upper_bound= Q3 + 1.5 * IQR

# Find outliers
outliers=(numerical_cols < lower_bound) | (numerical_cols > upper_bound )

outlier_values=numerical_cols[outliers]
print(outlier_values.stack())
```

```python
outlier_counts=outliers.sum()
print(outlier_counts)
```

## Skewness, kurtosis and Correlation

```python
from scipy.stats import skew, kurtosis
```

```python
skewness=numerical_cols.skew()
kurt=numerical_cols.kurtosis()
correlation=numerical_cols.corr()

print('Skewness: ',skewness)
print('\nKurtosis: ',kurt)
print('\nCorrelation: ',correlation)
```

```python
transformed_casual=dataset5['casual_log']=np.log1p(dataset5['casual'])
transformed_registered=dataset5['registered_log']=np.log1p(dataset5['registered'])
transformed_cnt=dataset5['cnt_log']=np.log1p(dataset5['cnt'])

print("Origional Skewness Casual: ",dataset5['casual'].skew())
print("\nTransformed Skewness Casual: ",transformed_casual.skew())
print("\nOrigional Skewness Registered: ",dataset5['registered'].skew())
print("\nTransformed Skewness Registered: ",transformed_registered.skew())
print("\nOrigional Skewness Cnt: ",dataset5['cnt'].skew())
print("\nTransformed Skewness Cnt: ",transformed_cnt.skew())
```
